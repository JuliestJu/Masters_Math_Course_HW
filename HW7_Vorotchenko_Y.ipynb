Розв'яжи програмно, згенерувавши всі можливі комбінації.

---
Розв'язання:

````python
import itertools
# Генерування всіх можливих комбінацій для трьох гральних кубиків
combinations = list(itertools.product(range(1, 7), repeat=3))

# Підрахунок комбінацій, що дають в сумі 11 і 12 очок
sum_11 = sum(sum(comb) == 11 for comb in combinations)
sum_12 = sum(sum(comb) == 12 for comb in combinations)

sum_11, sum_12
````

Кількість можливих комбінацій, що дають в сумі 11 очок, становить 27, тоді як комбінацій, що дають в сумі 12 очок, — 25.

## Завдання 7. Випадкове блукання
Ціна акцій компанії в кожен момент часу може з рівною ймовірністю збільшитись на 2 або зменшитись на 1. В початковий момент часу $t=0$ ціна рівна 0. Визнач середню ціну акції через 3 одиниці часу ($t=3$).  
а) Виріши аналітично.

б) Напиши симуляцію данного процесу. Порахуй середнє значення та намалюй гістограму ціни для 10, 100, 1000 та 10000 симуляцій.

---
Розв'язання:

image.png

````python
import numpy as np
import matplotlib.pyplot as plt

def stock_price_at_time(t):

    # Початкова ціна
    price = 0
    for _ in range(t):
        # Рандомне збільшення на 2 або зменшення на 1
        price += 2 if np.random.rand() < 0.5 else -1
    return price

def simulate_n_times(n, t):
    np.random.seed(42)
    prices = [stock_price_at_time(t) for _ in range(n)]
    mean_price = np.mean(prices)
    print(f"Середня ціна акції через {t} кроків при {n} симуляціях: {mean_price}")

    # Вивід гістограми
    plt.figure(figsize=(10, 6))
    plt.hist(prices, bins=20, color='blue', edgecolor='black', alpha=0.7)
    plt.title(f"Розподіл цін акцій через {t} кроків при {n} симуляціях")
    plt.xlabel("Ціна акції")
    plt.ylabel("Кількість симуляцій")
    plt.grid(axis='y', alpha=0.75)
    plt.show()
````

````python
import matplotlib.pyplot as plt
for n in [10, 100, 1000, 10000]:
    simulate_n_times(n, 3)
````
