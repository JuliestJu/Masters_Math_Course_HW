```python
import sympy as sp

x = sp.symbols('x')

f1 = x**3/3 + x**2/2 - 2*x
f2 = (x**2 + 1)**(1/2)
f3 = 1 /((x**2 + 1)**(1/2))

f1_diff = sp.diff(f1, x)
f2_diff = sp.diff(f2, x)
f3_diff = sp.diff(f3, x)

point1 = 1
point2 = -(1/2)

f1_diff_value_point1 = f1_diff.subs(x, point1)
f1_diff_value_point2 = f1_diff.subs(x, point2)

f2_diff_value_point1 = f2_diff.subs(x, point1)
f2_diff_value_point2 = f2_diff.subs(x, point2)

f3_diff_value_point1 = f3_diff.subs(x, point1)
f3_diff_value_point2 = f3_diff.subs(x, point2)

print("Похідна першої функції: ", f1_diff)
print("Похідна другої функції: ", f2_diff)
print("Похідна третьої функції: ", f3_diff)

print("Значення похідної першої функції у точці x=1: ", f1_diff_value_point1)
print("Значення похідної першої функції у точці x=-1/2: ", f1_diff_value_point2)
print("Значення похідної другої функції у точці x=1: ", f2_diff_value_point1)
print("Значення похідної другої функції у точці x=-1/2: ", f2_diff_value_point2)
print("Значення похідної третьої функції у точці x=1: ", f3_diff_value_point1)
print("Значення похідної третьої функції у точці x=-1/2: ", f3_diff_value_point2)
```

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Визначення символьних змінних та функції
x = sp.Symbol('x')
y = -3*x**2 + 30*x

# Обчислення похідної
y_prime = sp.diff(y, x)
print("Перша похідна: ", y_prime)

# Знаходження точок екстремуму
critical_points = sp.solve(y_prime, x)
print("Точки екстремуму: ", critical_points)

# Знаходження другої похідної для визначення типу екстремуму
y_double_prime = sp.diff(y_prime, x)
print("Друга похідна: ", y_double_prime)
second_derivative_values = [y_double_prime.subs(x, point) for point in critical_points]

# Вибираємо тільки ті точки, де друга похідна менше за 0 (для точок максимуму)
max_points = [point for point, value in zip(critical_points, second_derivative_values) if value < 0]

# Візуалізація
x_values = np.linspace(0, 10, 401)
y_values = [-3*x**2 + 30*x for x in x_values]
y_prime_values = [y_prime.subs(x, x_val) for x_val in x_values]

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='$y = -3x^2 + 30x$', color='blue')
plt.plot(x_values, y_prime_values, label="y'", color='red')
plt.scatter(critical_points, [0]*len(critical_points), color='black', label='Critical Points')
plt.scatter(max_points, [y.subs(x, point) for point in max_points], color='green', label='Maximum Points')

# Відображення точки максимуму
if max_points:
    max_point = max_points[0]
    plt.scatter(max_point, y.subs(x, max_point), color='red', label='Maximum Point')
    plt.annotate(f'Maximum: ({max_point}, {y.subs(x, max_point)})',
                 xy=(max_point, y.subs(x, max_point)), xytext=(max_point, y.subs(x, max_point) - 30),
                 arrowprops=dict(facecolor='black', arrowstyle='->'),
                 fontsize=10, color='black')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of $y = -3x^2 + 30x$ and its Derivative')
plt.legend()
plt.grid(True)
plt.show()
```
